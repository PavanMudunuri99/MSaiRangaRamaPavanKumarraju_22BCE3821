```python
# Generate Candidate Itemsets

def generate_candidates(transactions):
    candidates = set()
    for transaction in transactions:
        for item in transaction:
            candidates.add(frozenset([item])) 
    
    return candidates
candidate_itemsets = generate_candidates(transactions)
for itemset in candidate_itemsets:
    print(itemset)
```

```python
# Calculate Support for Itemsets

def calculate_support(itemsets, transactions):
    itemset_count = {itemset: 0 for itemset in itemsets}
    for transaction in transactions:
        for itemset in itemsets:
            if itemset.issubset(transaction):
                itemset_count[itemset] += 1
    total_transactions = len(transactions)
    support = {itemset: count / total_transactions for itemset, count in itemset_count.items()}
    
    return support

support_1_itemsets = calculate_support(candidate_itemsets, transactions)
for itemset, sup in support_1_itemsets.items():
    print(f"Itemset: {itemset}, Support: {sup}")
```

```python
# Generate Frequent Itemsets

def apriori(transactions, min_support):
    candidate_1_itemsets = generate_candidates(transactions)
    support_1_itemsets = calculate_support(candidate_1_itemsets, transactions)
    
    frequent_itemsets = {itemset: support for itemset, support in support_1_itemsets.items() if support >= min_support}
    k = 2
    while True:
        candidate_k_itemsets = generate_combinations(frequent_itemsets, k)
        support_k_itemsets = calculate_support(candidate_k_itemsets, transactions)
        frequent_k_itemsets = {itemset: support for itemset, support in support_k_itemsets.items() if support >= min_support}
        
        if not frequent_k_itemsets:
            break
        
        frequent_itemsets.update(frequent_k_itemsets)
        k += 1 
    
    return frequent_itemsets

def generate_combinations(frequent_itemsets, k):
    itemsets = list(frequent_itemsets.keys())
    candidates = set()
    
    for combo in combinations(itemsets, 2):
        union_set = combo[0].union(combo[1])
        if len(union_set) == k:
            candidates.add(union_set)
    
    return candidates

min_support = 0.03  
frequent_itemsets = apriori(transactions, min_support)
print(frequent_itemsets)  
```

```python
# Generate Association Rules

def generate_rules(frequent_itemsets, transactions, min_confidence):
    rules = []
    
    for itemset, support in frequent_itemsets.items():
        for size in range(1, len(itemset)):
            subsets = combinations(itemset, size)
            
            for subset in subsets:
                antecedent = frozenset(subset)
                consequent = itemset - antecedent
                
                antecedent_support = frequent_itemsets.get(antecedent, 0)
                rule_support = support
                confidence = rule_support / antecedent_support if antecedent_support != 0 else 0
                
                if confidence >= min_confidence:
                    rules.append((antecedent, consequent, confidence))
    
    return rules

min_confidence = 0.5 
rules = generate_rules(frequent_itemsets, transactions, min_confidence)
print(rules) 
```

```python
# Building the FP-Tree

from collections import defaultdict

class FPNode:
    def __init__(self, item, count=0, parent=None):
        self.item = item
        self.count = count
        self.parent = parent
        self.children = {}

def build_fp_tree(transactions, min_support):
    item_counts = defaultdict(int)
    for transaction in transactions:
        for item in transaction:
            item_counts[item] += 1

    item_counts = {item: count for item, count in item_counts.items() if count >= min_support}
    
    root = FPNode(None)  
    for transaction in transactions:
        sorted_items = [item for item in transaction if item in item_counts]
        sorted_items.sort(key=lambda item: item_counts[item], reverse=True)
        current_node = root
        for item in sorted_items:
            if item not in current_node.children:
                current_node.children[item] = FPNode(item, 1, current_node)
            else:
                current_node.children[item].count += 1
            current_node = current_node.children[item]

    return root, item_counts
min_support = 3  
fp_tree, item_counts = build_fp_tree(transactions, min_support)
print("FP-Tree built with frequent itemsets:", item_counts)

```

```python
# Mining the FP-Tree

def mine_fp_tree(fp_tree, min_support, prefix=[]):
    frequent_itemsets = []
    for item, node in fp_tree.children.items():
        item_support = node.count
        if item_support >= min_support:
            frequent_itemsets.append((prefix + [item], item_support))
            conditional_pattern_base = generate_conditional_pattern_base(node)

            conditional_fp_tree, _ = build_fp_tree(conditional_pattern_base, min_support)
            frequent_itemsets.extend(mine_fp_tree(conditional_fp_tree, min_support, prefix + [item]))

    return frequent_itemsets

def generate_conditional_pattern_base(node):
    conditional_pattern_base = []
    while node.parent is not None:
        path = []
        current_node = node
        while current_node.parent is not None:
            path.append(current_node.item)
            current_node = current_node.parent
        path.reverse()  
        conditional_pattern_base.append(path)
        node = node.parent
    return conditional_pattern_base

frequent_itemsets = mine_fp_tree(fp_tree, min_support)
print("Frequent itemsets mined from the FP-tree:", frequent_itemsets)
```

```python
# Apriori Algorithm

import pandas as pd
import itertools
from collections import defaultdict

def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path)
    transactions = df['Items'].apply(lambda x: [item.strip() for item in x.split(',')]).tolist()  # Strip whitespace
    return transactions

# Generate Candidate Itemsets
def generate_candidates(transactions, k, previous_frequent_itemsets=None):
    if k == 1:
        candidates = set()
        for transaction in transactions:
            for item in transaction:
                candidates.add(frozenset([item]))
        return candidates
    else:
        candidates = set()
        for itemset1 in previous_frequent_itemsets:
            for itemset2 in previous_frequent_itemsets:
                union = itemset1.union(itemset2)
                if len(union) == k:
                    candidates.add(union)
        return candidates
def calculate_support(candidates, transactions):
    itemset_counts = defaultdict(int)
    for transaction in transactions:
        transaction_set = set(transaction)  
        for candidate in candidates:
            if candidate.issubset(transaction_set):
                itemset_counts[candidate] += 1
    num_transactions = len(transactions)
    itemset_support = {itemset: count / num_transactions for itemset, count in itemset_counts.items()}
    return itemset_support

# Prune Infrequent Itemsets
def prune_infrequent_itemsets(itemset_support, min_support):
    frequent_itemsets = {itemset: support for itemset, support in itemset_support.items() if support >= min_support}
    return frequent_itemsets
def generate_association_rules(frequent_itemsets, min_confidence):
    rules = []
    for itemset in frequent_itemsets:
        if len(itemset) > 1:
            for i in range(1, len(itemset)):
                antecedents = itertools.combinations(itemset, i)
                for antecedent in antecedents:
                    antecedent = frozenset(antecedent)
                    consequent = itemset - antecedent
                    confidence = frequent_itemsets[itemset] / frequent_itemsets[antecedent]
                    if confidence >= min_confidence:
                        rules.append((antecedent, consequent, confidence))
    return rules

def apriori(file_path, min_support=0.2, min_confidence=0.5):
    transactions = load_and_preprocess_data(file_path)

    k = 1
    frequent_itemsets = {} 
    while True:
        candidates = generate_candidates(transactions, k, frequent_itemsets.get(k - 1)) 
        itemset_support = calculate_support(candidates, transactions)
        current_frequent_itemsets = prune_infrequent_itemsets(itemset_support, min_support)

        if not current_frequent_itemsets:
            break

        frequent_itemsets[k] = current_frequent_itemsets 
        k += 1

    all_frequent_itemsets = {}
    for itemsets in frequent_itemsets.values():
        all_frequent_itemsets.update(itemsets)

    association_rules = generate_association_rules(all_frequent_itemsets, min_confidence)
    return all_frequent_itemsets, association_rules

file_path = 'Book1.csv'  
min_support = 0.2       
min_confidence = 0.5   

frequent_itemsets, association_rules = apriori(file_path, min_support, min_confidence)

print("Frequent Itemsets:")
for itemset, support in frequent_itemsets.items():
    print(f"{set(itemset)}: {support:.2f}")

print("\nAssociation Rules:")
for antecedent, consequent, confidence in association_rules:
    print(f"{set(antecedent)} -> {set(consequent)}: {confidence:.2f}")
```

```python
# FP-Algorithm for the Book.csv file

import pandas as pd
from collections import defaultdict
from itertools import combinations

# Load the data and preprocess it into a list of transactions
def load_data(file_path):
    df = pd.read_csv(file_path)  # Correct indentation
    transactions = df['Items'].apply(lambda x: x.split(',')).tolist()  # Correct indentation
    print("Transactions Loaded:", transactions)  # Correct indentation
    return transactions  # Correct indentation

# Create the FP-Tree and mine frequent itemsets
class FPTree:
    def __init__(self, min_support=0.3):
        self.min_support = min_support
        self.header_table = defaultdict(int)
        self.tree = {}
        self.frequent_itemsets = []

    def build_tree(self, transactions):
        for transaction in transactions:
            for item in transaction:
                self.header_table[item] += 1

        self.header_table = {item: count for item, count in self.header_table.items() if count / len(transactions) >= self.min_support}
        print("Header Table:", self.header_table)

        for transaction in transactions:
            sorted_items = [item for item in transaction if item in self.header_table]
            sorted_items.sort(key=lambda x: self.header_table[x], reverse=True)
            if sorted_items:
                self.tree = self.insert_tree(sorted_items, self.tree)

    def insert_tree(self, items, tree):
        if not items:
            return tree
        current_item = items[0]
        if current_item in tree:
            tree[current_item][0] += 1
        else:
            tree[current_item] = [1, {}]
        tree[current_item][1] = self.insert_tree(items[1:], tree[current_item][1])
        return tree

    def mine_frequent_itemsets(self, transactions):
        # Traverse the FP-tree and extract itemsets
        for item, (count, subtree) in self.tree.items():
            support = count / len(transactions)  # Calculate support
            if support >= self.min_support:
                self.frequent_itemsets.append(([item], count))
            self.mine_subtree(subtree, [item], count, transactions)
        print("Frequent Itemsets Found:", self.frequent_itemsets)
        return self.frequent_itemsets

    def mine_subtree(self, subtree, prefix, prefix_count, transactions):
        if not subtree:
            return
        for item, (count, subsubtree) in subtree.items():
            support = count / len(transactions)
            if support >= self.min_support:
                self.frequent_itemsets.append((prefix + [item], count))
                self.mine_subtree(subsubtree, prefix + [item], count, transactions)


def generate_association_rules(frequent_itemsets, transactions, min_confidence=0.7):
    rules = []
    itemsets_dict = defaultdict(int)

    for itemset, count in frequent_itemsets:
        itemsets_dict[tuple(itemset)] = count / len(transactions)

    for itemset, count in frequent_itemsets:
        if len(itemset) > 1:
            # Generate possible antecedents
            for length in range(1, len(itemset)):
                antecedents = combinations(itemset, length)
                for antecedent in antecedents:
                    antecedent_tuple = tuple(antecedent)
                    consequent_tuple = tuple(set(itemset) - set(antecedent))
                    if antecedent_tuple in itemsets_dict:
                        confidence = itemsets_dict[tuple(itemset)] / itemsets_dict[antecedent_tuple]
                        if confidence >= min_confidence:
                            rules.append((list(antecedent), list(consequent_tuple), confidence))
    return rules

def fp_growth_algorithm(file_path, min_support=0.3, min_confidence=0.7):
    transactions = load_data(file_path)

    fptree = FPTree(min_support=min_support)
    fptree.build_tree(transactions)
    frequent_itemsets = fptree.mine_frequent_itemsets(transactions)
    association_rules_result = generate_association_rules(frequent_itemsets, transactions, min_confidence)
    print("Frequent Itemsets:")
    for itemset, count in frequent_itemsets:
        support = count / len(transactions)
        print(f"Itemset: {itemset}, Support: {support:.2f}")

    print("\nAssociation Rules:")
    if association_rules_result:
        for antecedent, consequent, confidence in association_rules_result:
            print(f"Rule: {antecedent} -> {consequent}, Confidence: {confidence:.2f}")
    else:
        print("No association rules generated")

fp_growth_algorithm('Book1.csv', min_support=0.3, min_confidence=0.7)
```
